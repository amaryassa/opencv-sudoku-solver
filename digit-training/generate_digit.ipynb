```python
from PIL import ImageFont, ImageDraw, Image
import numpy as np
import os
import cv2
from sklearn.utils import shuffle
```

```python
nbr=2*42
imagesLen=0
taille_image=32
```

```python
def modif_image(image, seuil=1):
    b=np.random.normal(0, 1, (taille_image, taille_image))
    a=image.copy()
    a[b>seuil]=255
    a[b<-seuil]=0
    return a
```

```python
def save_images(image_list, directory):
    if not os.path.exists(directory):
        os.makedirs(directory)
    # Vérifier le contenu du dossier pour déterminer le numéro d'incrémentation
    existing_files = os.listdir(directory)
    increment = len(existing_files)

    for idx, img in enumerate(image_list):
        filename = f"image_{increment + idx}.png"  # Nom du fichier avec l'incrémentation
        filepath = os.path.join(directory, filename)
        cv2.imwrite(filepath, img)
        #print(f"Image sauvegardée : {filename}")
```
